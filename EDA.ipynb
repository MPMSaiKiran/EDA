```python
# Dividing the dataset into two dataset of  target=1(client with payment difficulties) and target=0(all other)

target0_df=df.loc[df["TARGET"]==0]
target1_df=df.loc[df["TARGET"]==1]
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('white')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 35
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.xscale('linear')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='magma') 
        
    plt.show()
```

```python
uniplot(target1_df,col='NAME_EDUCATION_TYPE',title='defaulters')
uniplot(target0_df,col='NAME_EDUCATION_TYPE',title='non_defaulters')
```

```python
uniplot(target0_df,col='NAME_INCOME_TYPE',title='non_defaulters')
uniplot(target1_df,col='NAME_INCOME_TYPE',title='defaulters')
```
